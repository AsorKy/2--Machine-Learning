###3.3-Data Scaling

We standarize our data and delete the variable "country" which is the name of the country.

```python
from sklearn.preprocessing import StandardScaler
scaaler = StandardScaler()
df_country_scaled = scaaler.fit_transform(df_country.drop('country',axis=1))
```

```python
df_country_scaled = pd.DataFrame(df_country_scaled, columns = df_country.drop('country',axis=1).columns )
```

```python
df_country_scaled
```

###3.4-PCA dimensionality reduction

To ensure better convergence and possible visualization, we apply Principal Component Analysis in the standarized dataset.

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
pca.fit(df_country_scaled)
pca_data_scaled = pca.transform(df_country_scaled)
pca_data_scaled
```

To determine how much dimensionality reduction we are goint to use, we extract the variance ratio of the trained PCA which tell us how much of the data variance is explained by each of the components. Once these variances are determinated, we accumlate them to see untill which component, more of the 90% of the data is explained.

```python
var = pca.explained_variance_ratio_
print(var)
```

```python
cum_var = np.cumsum(np.round(var, decimals=4)*100)
plt.figure(figsize=(4,3))
plt.plot(cum_var, 'r-x')
plt.show()
```

The previous plot tell us that almost 90% of the data is explained if we proyect it into its first four principal components, then, we reduce the 8 variables to 4.

```python
pca_data_standard = pd.DataFrame(pca_data_scaled)
# We drop the pca data after the fourth component
pca_data_standard.drop([4,5,6,7,8],axis=1, inplace=True)
```

```python
pca_data_standard
```

##4)-Clustering Implementation

First, we import the needed libraries for clustering.

```python
# Clustering algorithms
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN

# Metric for performance
from sklearn.metrics import silhouette_score

# Metric for hierarchical clustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

###4.1- K-means clustering

####4.1.1- K-means with PCA

We apply K-means for the standarized PCA data for a number of clusters between 2 and 15.

```python
Sum_of_squared_distances = []
silhouette_scores = []
K = range(2,15)
for _ in K:
    km = KMeans(n_clusters = _)
    y = km.fit_predict(pca_data_standard)
    Sum_of_squared_distances.append(km.inertia_)
    silhouette_scores.append(silhouette_score(pca_data_standard,y))
```

```python
fig = plt.figure(figsize=(4,4))
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('K')
plt.ylabel('Inertia')
plt.show()
```

```python
fig = plt.figure(figsize=(4,4))
plt.plot(K, silhouette_scores, 'rx-')
plt.xlabel('K')
plt.ylabel('silhouette score')
plt.show()
```

```python
km = KMeans(n_clusters = 4)
y = km.fit_predict(pca_data_standard)
print("The silhuete score is=",silhouette_score(pca_data_standard, y))
df_country['k_means_pca'] = y
```

###4.3-DBSCAN clustering

####4.3.1-Dbscan with PCA

```python
from sklearn.neighbors import NearestNeighbors

# We first use k-neighbors to calculate an estimative of the epsilon-radious
neighbors = NearestNeighbors(n_neighbors=2)
neighbors_fit = neighbors.fit(pca_data_standard)
distances, indices = neighbors_fit.kneighbors(pca_data_standard)
```

```python
distances = np.sort(distances, axis=0)
distances = distances[:,1]
fig = plt.figure(figsize=(6,6))
plt.plot(distances)
```

We estimate a knee between 0.5 and 2.

```python
eps_values = np.arange(0.5, 2.0, 0.10)
min_samples = np.arange(3,12)
```

Having an estimate of the optimal knee, we make a combinatorics of all posible epsilon and number of samples pairs to check which has the highest silhouete score.

```python
from itertools import product
dbscan_paramns = list(product(eps_values,min_samples))
no_of_clusters = []
sil_score = []
for p in dbscan_paramns:
    y_dbscan = DBSCAN(eps=p[0], min_samples=p[1]).fit_predict(pca_data_standard)
    try:
        sil_score.append(silhouette_score(pca_data_standard,y_dbscan))
    except:
        sil_score.append(0)
    no_of_clusters.append(len(np.unique(y_dbscan)))
```

```python
df_param_tunning = pd.DataFrame.from_records(dbscan_paramns, columns=['Eps','Min_samples'])
df_param_tunning['sil_score'] = sil_score
df_param_tunning['n_clusters'] = no_of_clusters
```

```python
df_param_tunning.head()
```

```python
pivot_1 = pd.pivot_table(df_param_tunning, values='sil_score', columns='Eps', index='Min_samples')
pivot_2 = pd.pivot_table(df_param_tunning, values='n_clusters', columns='Eps', index='Min_samples')
```

```python
fig, ax = plt.subplots(figsize=(18,6))
sns.heatmap(pivot_1, annot=True, annot_kws={'size':10}, cmap='coolwarm', ax=ax)
plt.show()
```

Now we visualize how much clusters are formed for each combination.

```python
fig, ax = plt.subplots(figsize=(18,6))
sns.heatmap(pivot_2, annot=True, annot_kws={'size':10}, cmap='coolwarm', ax=ax)
plt.show()
```

According to the two previous heatmaps, the highest silhouete score (0.53) is achived for a two cluster segmentation. Despite the fact that the silhouete score is good, two clusters could mean have just one cluster and another category atributed to the noisi datapoints.
Therefore, qe need to seek the next highest cluster value (3) with the highest silhouete score (0.19) to consider it as the best candidate.

```python
dbscan_train = DBSCAN(eps=1.18, min_samples=3)
y_dbscan = dbscan_train.fit_predict(pca_data_standard)
print("he silhouete score is = ",silhouette_score(pca_data_standard,y_dbscan ))
df_country['dbscan_pca'] = y_dbscan
```
