##3.1-Systematic sampling

In inferential statistics, systematic sampling is a probability sampling method where elements from a population are selected at regular intervals. It involves selecting every kth element from the population, where k is a predetermined constant.



1.   Determine the target population: Define the population of interest from which you want to draw a sample.
2.   Define the sample size: Decide on the desired sample size, denoted as "n".
3.   Calculate the sampling interval: Divide the population size (N) by the sample size (n) to obtain the sampling interval (k). The sampling interval represents the number of elements between each selected element.
4.   Randomly select a starting point: Choose a random number between 1 and k as the starting point.
5.   Select sample elements: Begin at the randomly selected starting point and select every kth element from the population until the desired sample size is reached.



```python
def sys_sampling(dataframe,step,start=0):
  indexes = np.arange(start, len(dataframe), step=step)
  systematic_sample = dataframe.iloc[indexes]
  return systematic_sample
```

```python
systematic_sample = sys_sampling(econdata,3,0)
systematic_sample 
```

##3.3-Stratified sampling

In inferential statistics, stratified sampling is a probability sampling method that involves dividing the population into distinct subgroups or strata based on certain characteristics. The goal is to ensure that each subgroup is well-represented in the sample to obtain more accurate and precise estimates of population parameters.

The process of stratified sampling involves the following steps:


1.   Identify the target population: Define the population of interest from which you want to draw a sample.
2.   Determine relevant stratification factors: Identify the key characteristics or variables that are important for dividing the population into homogeneous subgroups or strata. These variables should be related to the research question or the purpose of the study.
3.   Divide the population into strata: Divide the population into mutually exclusive and collectively exhaustive strata based on the identified stratification factors. Each element in the population should belong to only one stratum.
4.   Determine the sample size: Decide on the desired sample size for each stratum. The sample size can be proportional to the size of each stratum or can be allocated based on the variability within each stratum.
5.   Randomly select samples from each stratum: Independently and randomly select samples from each stratum using any appropriate sampling technique (such as simple random sampling or systematic sampling).
6.   Combine the stratum samples: Once the samples from each stratum are obtained, combine them to form the final stratified sample.

Stratified sampling is advantageous because it ensures that each subgroup or stratum is adequately represented in the sample, allowing for more precise estimates of population parameters for each subgroup. This method is particularly useful when there are substantial differences or variations between the subgroups in the population. It helps to reduce sampling variability and can provide more reliable and accurate results compared to other sampling techniques.

Now, lets imagine that an external source of data says to us that the proportions of data described by each element class of the new element should be:


1.   Hotels in Cuauhtemoc = 0.5
2.   Museums in Cuauhtémoc = 0.2
3.   Hotels in Venustiana Carranza = 0.1
4.   Markets in Cuauhtémoc = 0.1
5.   Markets in Venustiana Carranza = 0.1

Then, we aregoint to create this relection for the total population such taht our sample has the proportions that we want.


```python
def stratified_data(data, nombres_columnas_estrat, valores_estrat, prop_estrat, random_state=None):
    
    df_estrat = pd.DataFrame(columns = data.columns) # Creamos un data frame vacío con los nombres de las columnas de econdata

    pos = -1
    for i in range(len(valores_estrat)): # iteración sobre los valores estratificados
        pos += 1
        if pos == len(valores_estrat) - 1: 
            ratio_len = len(data) - len(df_estrat) # si es la iteración final calcula el número de valores de salida tenga el mismo número de filas que de entrada
        else:
            ratio_len = int(len(data) * prop_estrat[i]) # calcula el número de filas según la proporción deseada

        df_filtrado = econdata[data[nombres_columnas_estrat] ==valores_estrat[i]] # filtra los datos de origen según los valores seleccionados en la estratificación de datos
        df_temp = df_filtrado.sample(replace=True, n=ratio_len, random_state=random_state) # haz un sample de los datos filtrados usando la ratio que hemos calculado
        
        df_estrat = pd.concat([df_estrat, df_temp]) # junta las tablas de sample con la estratificada para producir el resultado final
        
    return df_estrat # Return the stratified, re-sampled data   
```

```python
strat_vals = ['Cuautémoc,Hotel', 'Cuautémoc,Museo', 'Venustiano Carranza,Hotel','Cuauhtémoc,Mercado','Venustiano Carranza,Mercado']
prop_strat = [0.5,0.2,0.1,0.1,0.1]
df_strat = stratified_data(econdata, 'estratificado', strat_vals, prop_strat, random_state=42)
```
