#1)-Library importing

```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as st
```

#2)-Data creation

For this excercise, we create a random distributed data.

```python
np.random.seed(20)
var1 = np.random.randint(0,10,10) # vacation days
var2 = var1 + np.random.normal(0,1,10) # spended money
```

#3)-Data visualization

In this excercise, we are going to visualize the data within an 80% confidence interval.

```python
df = pd.DataFrame({'days':var1,'spended_money':var2})
grafico = sns.regplot(x='days',y='spended_money', data=df, ci=80)
```

```python
tuple = list(zip(var1,var2))
tuple
```

```python
tabla = pd.DataFrame(tuple,columns=["Days","Spended_Money"])
tabla.head()
```

#3)-Interval visualization

```python
st.t.interval(alpha=0.95,
              df=len(tabla)-1,
              loc=np.mean(tabla),
              scale=st.sem(tabla))
```

If we’re working with larger samples (n≥30), we can assume that the sampling distribution of the sample mean is normally distributed (thanks to the Central Limit Theorem) and can instead use the norm.interval() function from the scipy.stats library.

The following example shows how to calculate a confidence interval for the true population mean height (in inches) of a certain species of plant, using a sample of 50 plants:

```python
import numpy as np
import scipy.stats as st

#define sample data
np.random.seed(0)
data = np.random.randint(10, 30, 50)

#create 95% confidence interval for population mean weight
st.norm.interval(alpha=0.95, loc=np.mean(data), scale=st.sem(data))

(17.40, 21.08)
```
