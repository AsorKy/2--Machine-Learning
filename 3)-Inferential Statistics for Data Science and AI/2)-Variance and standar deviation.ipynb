#1)-Library importing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#2)-Data importing

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
names = ["sepal-length","sepal-width","path-length","path-width","class"]
iris = pd.read_csv(url, names = names)
```

```python
iris.head()
```

#3)-Data Exploration

```python
plt.hist(iris["sepal-length"], color = "orange", bins = int(180/5))
plt.show()
```

```python
sns.distplot(iris['sepal-length'], hist=True, kde=True,
             bins=int(185/5), color='blue')
plt.show()
```

Now that we have check the normal nature of our variable of interest, we can check the variance of the variable.

```python
irisvar = iris['sepal-length'].var()
irisvar 
```

Now that we have check the normal nature of our variable of interest, we can check the standard deviation of the variable.

```python
irisstd = iris['sepal-length'].std()
irisstd
```

To check wether if the standar deviation or the variance is high, we need to compare it to the mean value of the distirbution.

```python
irismean = iris['sepal-length'].mean()
irismean
```

```python
print("the proportion covered by the variance is = ", irisvar/irismean )
print("the proportion covered by the std is = ", irisstd/irismean )
```

#4)-Data sampling

```python
sample = iris.sample(frac=0.5)
sample
```

```python
sample_var = sample['sepal-length'].var()
sample_std = sample['sepal-length'].std()
sample_mean = sample['sepal-length'].mean()

print("The variance of the sample is = {}".format(sample_var))
print("The std of the sample is = {}".format(sample_std))
print("The mean of the sample is = {}".format(sample_mean))
```

```python
print("the proportion covered by the variance is = ", sample_var/sample_mean )
print("the proportion covered by the std is = ", sample_std/sample_mean )
```
