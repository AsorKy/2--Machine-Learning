#2)-Spectral matrix decomposition

```python
A = np.array([[3,2],
              [4,1]])
print(A)
```

```python
eigenvalues, eigenvectors = np.linalg.eig(A)
print(eigenvalues)
print(eigenvectors)
```

```python
V = eigenvectors
lamb = np.diag(eigenvalues)
I_v = np.linalg.inv(V)
```

```python
A_calc = V.dot(lamb).dot(I_v)
print(A_calc)
```

A must be a real an symmetric matrix.

```python
A = np.array([[3,2],
              [2,3]])
print(A)
print(A==A.T)
```

```python
lamb, V = np.linalg.eig(A)
diag_lamb = np.diag(lamb)
print(V)
print(lamb)
```
