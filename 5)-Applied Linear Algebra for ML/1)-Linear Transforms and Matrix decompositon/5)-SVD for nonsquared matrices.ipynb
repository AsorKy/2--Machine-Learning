#2)-SVD for non-squared matrices

1. U must be an orthonormal matrix where its columns are the singular left vectors of the original matriz.
2. V must be a orthonormal matrix where its columns are the singular right  vectors of the original matrix.
3. D is a diagonal matrix composed of the calculated singular values.

```python
A = np.array([[1,2,3],
              [3,4,5]])
print(A)
```

```python
U, D, V = np.linalg.svd(A)
D = np.diag(D)
print(U)
print(D)
print(V)
```

```python
A_cal = U.dot(D.dot(V))
print(A_cal)
```
