```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

#2)-Data generation

```python
np.random.seed(42)

x = 3*np.random.rand(200)
y = 20*x + 2*np.random.randn(200)

x = x.reshape(200,1)
y = y.reshape(200,1)

print(x[0:10])
```

```python
xy = np.hstack([x,y])
print(xy.shape)
```

```python
plt.plot(xy[:,0],xy[:,1],'.')
plt.show()
```

```python
xy_centrado = xy - np.mean(xy, axis = 0)
plt.plot(xy_centrado[:,0],xy_centrado[:,1],'.')
plt.show()
```

```python
eigenvalues, eigenvectors = np.linalg.eig(xy_centrado.T.dot(xy_centrado))
print(eigenvectors)
```
