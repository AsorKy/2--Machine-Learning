```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

#2)-Image compression using SVD

```python
path = '/content/drive/MyDrive/Colab Notebooks/2)-Machine Learning/2)-Platzy Career (Deep Learning)/2)-Convolutional Neural Networks with TensorFlow and Keras/data/vishnu_1.jpeg'
imagen = Image.open(path)
plt.imshow(imagen)
plt.show()
```

```python
imagen_gr = imagen.convert('LA')
imagen_mat = np.array(list(imagen_gr.getdata(band=0)),float)
imagen_mat.shape = (imagen_gr.size[1], imagen_gr.size[0])
print(imagen_mat)
```

```python
U, D, V = np.linalg.svd(imagen_mat)
```

```python
print(imagen_mat.shape)
print(U.shape)
print(D.shape)
print(V.shape)
```

```python
imagen_recons = np.matrix(U[:,:1]) * np.diag(D[:1]) * np.matrix(V[:1,:])
plt.imshow(imagen_recons, cmap='gray')
plt.show()
```

```python
i = 5
imagen_recons = np.matrix(U[:,:i]) * np.diag(D[:i]) * np.matrix(V[:i,:])
plt.imshow(imagen_recons, cmap='gray')

titulo = 'Valores singultaes = %s' % i
plt.title(titulo)
plt.show()
```
