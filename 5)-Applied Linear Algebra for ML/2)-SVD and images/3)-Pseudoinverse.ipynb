#2)-Pseudoinverse of a matrix


```python
A = np.array([[2,3],[5,7],[11,13]])
print(A)
```

```python
U, D, V = np.linalg.svd(A)
print(U)
print('U')
print('D')
print(D)
print('V')
print(V)
```

```python
D_pse = np.zeros((A.shape[0], A.shape[1])).T
print(D_pse)
```

```python
print('Valores a remplazar en D_pse')
print(D_pse[:D.shape[0], :D.shape[0]])
```

```python
print('Valores que pondremos en D_pse')
print(np.linalg.inv(np.diag(D)))
```

```python
print('D_pse')
D_pse[:D.shape[0], :D.shape[0]] = np.linalg.inv(np.diag(D))
print(D_pse)
```

```python
A_pse = V.T.dot(D_pse).dot(U.T)
print(A_pse)
```

```python
A_pse_calc = np.linalg.pinv(A)
print(A_pse_calc)
```

```python
print(A_pse.dot(A))
```

The pseudoinverse is an approximation to the real matrix inverse.

```python
np.set_printoptions(suppress=False)
```

```python
print(A_pse.dot(A))
```

The pseudoinverse is not conmutative

```python
print(A.dot(A_pse))
```

Another way to calculate the pseudoinverse of a matrix without using the SVD method at the cost of being more  inaccurate in some scenarios, is the following.

```python
A_pse_2 = np.linalg.inv(A.T.dot(A)).dot(A.T)
print(A_pse_2)
```
