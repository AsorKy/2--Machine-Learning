```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

#2)-Images as matrices

```python
plt.style.use('classic')

path = '/content/drive/MyDrive/Colab Notebooks/2)-Machine Learning/2)-Platzy Career (Deep Learning)/2)-Convolutional Neural Networks with TensorFlow and Keras/data/vishnu_1.jpeg'
imagen = Image.open(path)
plt.imshow(imagen)
plt.show()
```

```python
imagen_gr = imagen.convert('LA')
print(imagen_gr)
```

```python
plt.imshow(imagen_gr)
plt.show()
```

```python
imagen_mat = np.array(list(imagen_gr.getdata(band=0)),float)
print(imagen_mat)
```

```python
imagen_mat.shape = (imagen_gr.size[1], imagen_gr.size[0])
print(imagen_mat)
```

```python
print(imagen_mat.shape)
```

```python
plt.imshow(imagen_mat, cmap = 'gray')
plt.show()
```

```python
imagen_mat_2 = imagen_mat/10
print(imagen_mat_2)
```

```python
plt.imshow(imagen_mat_2)
plt.show()
```

```python
plt.imshow(imagen_mat_2, cmap='gray')
plt.show()
```

```python
Max_1 = np.max(imagen_mat)
Max_2 = np.max(imagen_mat_2)
Min_1 = np.min(imagen_mat)
Min_2 = np.min(imagen_mat_2)
print(Max_1)
print(Max_2)
print(Min_1)
print(Min_2)
```

```python
imagen_mat_3 = imagen_mat / Max_1
plt.imshow(imagen_mat_3)
plt.show()
```
